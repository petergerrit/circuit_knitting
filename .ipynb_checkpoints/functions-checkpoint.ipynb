```python
import numpy as np
```

```python
"""Convert a tuple representing a Cartesian vector to a tuple representing the vector in spherical coordinates"""

def cartesian_to_spherical(cart_vec):
    x, y, z = cart_vec
    rad = np.sqrt((cart_vec**2).sum())
    theta = np.arctan2(np.sqrt(x**2+y**2), z)
    phi = np.arctan2(y, x)
    return rad, theta, phi
```

```python
"""Compute the g-factor for a given set of four three-momenta"""

# g factor for H_vv
def g_factor(p1, p2, q1, q2):
    absp1, tp1, pp1 = cartesian_to_spherical(p1)
    absp2, tp2, pp2 = cartesian_to_spherical(p2)
    absq1, tq1, pq1 = cartesian_to_spherical(q1)
    absq2, tq2, pq2 = cartesian_to_spherical(q2)
    fac1 = np.exp(-1j*pq1)*np.sin(tq1/2.)*np.cos(tq2/2.) - np.exp(-1j*pq2)*np.cos(tq1/2.)*np.sin(tq2/2.)
    fac2 = np.exp(1j*pp1)*np.sin(tp1/2.)*np.cos(tp2/2.) - np.exp(1j*pp2)*np.cos(tp1/2.)*np.sin(tp2/2.)
    return 2 * fac1 * fac2
```
