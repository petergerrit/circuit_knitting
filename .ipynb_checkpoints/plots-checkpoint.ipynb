```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shots_list = [1024*2**i for i in range(11)]
```

# 1st Trotter step, epsilon 0.8, no noise, no knit

```python
ferm_num_step1_epsilon08_no_noise_no_knit = []
with open('data/ferm_num_step1_epsilon08_no_noise_no_knit.txt', 'r') as file:
    for line in file:
        val = line[:-1]
        ferm_num_step1_epsilon08_no_noise_no_knit.append(float(val))

bs_error_step1_epsilon08_no_noise_no_knit = []
with open('data/bs_error_step1_epsilon08_no_noise_no_knit.txt', 'r') as file:
    for line in file:
        val = line[:-1]
        bs_error_step1_epsilon08_no_noise_no_knit.append(float(val))
```

```python
plt.scatter(shots_list, ferm_num_step1_epsilon08_no_noise_no_knit)
plt.errorbar(shots_list, ferm_num_step1_epsilon08_no_noise_no_knit, bs_error_step1_epsilon08_no_noise_no_knit, alpha=0.2, ls='none')
plt.xscale('log')
plt.xlabel('shots')
plt.ylabel('mean fermion number')
plt.title('1st Trotter step: not knitted, no noise')
plt.show()
```
