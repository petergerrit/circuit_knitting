```python
import import_ipynb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pickle

from functions import *
from params import *
```

```python
shots_list = [1024*i**2 for i in range(11)]
ferm_num_step1_epsilon08_no_noise = []
bs_error_step1_epsilon08_no_noise = []
for i in range(11):
    with open('data/step1_epsilon08_count' + str(1024*2**i) + '_no_noise_knitted.pkl', 'rb') as file:
        counts = pickle.load(file)
    ferm_num_step1_epsilon08_no_noise.append(fermion_number(counts, mid))
    bs_error_step1_epsilon08_no_noise.append(bootstrap_error(counts, mid, 1024*2**i))
```

```python
ferm_num_step1_epsilon08_with_noise = []
bs_error_step1_epsilon08_with_noise = []
for i in range(11):
    with open('data/step1_epsilon08_count' + str(1024*2**i) + '_with_noise_knitted.pkl', 'rb') as file:
        counts = pickle.load(file)
    ferm_num_step1_epsilon08_with_noise.append(fermion_number(counts, mid))
    bs_error_step1_epsilon08_with_noise.append(bootstrap_error(counts, mid, 1024*2**i))
```

```python
plt.scatter(shots_list, ferm_num_step1_epsilon08_no_noise)
plt.errorbar(shots_list, ferm_num_step1_epsilon08_no_noise, bs_error_step1_epsilon08_no_noise, alpha=0.2, ls='none')
plt.xscale('log')
plt.xlabel('shots')
plt.ylabel('mean fermion number')
plt.title('1st Trotter step: knitted, no noise')
plt.show()
```

```python
plt.scatter(shots_list, ferm_num_step1_epsilon08_with_noise, marker = 'v')
plt.errorbar(shots_list, ferm_num_step1_epsilon08_with_noise, bs_error_step1_epsilon08_with_noise, alpha=0.2, ls='none')
plt.xscale('log')
plt.xlabel('shots')
plt.ylabel('mean fermion number')
plt.title('1st Trotter step: knitted, with noise')
plt.show()
```
