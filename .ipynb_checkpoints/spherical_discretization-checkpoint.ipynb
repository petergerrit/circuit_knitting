```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rho1 = 3
phi1 = 3*2*np.pi/9
theta1 = 5*2*np.pi/9
```

```python
rho2 = 1
phi2 = 4*2*np.pi/9
theta2 = 8*2*np.pi/9
```

```python
def pol_to_car(rho, theta, phi):
    x = rho*np.cos(theta)*np.sin(phi)
    y = rho*np.sin(theta)*np.sin(phi)
    z = rho*np.cos(phi)
    return x, y, z
```

```python
def car_to_pol(x, y, z):
    rho = np.sqrt(x**2 + y**2 + z**2)
    theta = np.arctan2(y, x)
    phi = np.arctan2(np.sqrt(rho**2 - z**2), z)
    return rho, np.mod(theta, 2*np.pi), np.mod(phi, np.pi)
```

```python
x1, y1, z1 = pol_to_car(rho1, theta1, phi1)
```

```python
x2, y2, z2 = pol_to_car(rho2, theta2, phi2)
```

```python
car_to_pol(x1+x2, y1+y2, z1+z2)
```

```python
# x3, y3, z3 = pol_to_car(1, 2*np.pi / 3, np.pi/2)
x3, y3, z3 = pol_to_car(2, 0, np.pi/2)
```

```python
x4, y4, z4 = pol_to_car(1, 4*np.pi / 3, np.pi/2)
```

```python
car_to_pol(x3+x4, y3+y4, z3+z4)
```

```python
x3+x4, y3+y4, z3+z4
```

```python
x4, y4
```

```python
plt.plot([np.cos(i) for i in np.linspace(0, 2*np.pi, 100)], [np.sin(i) for i in np.linspace(0, 2*np.pi, 100)], color='black', linestyle='dotted')
plt.plot([2*np.cos(i) for i in np.linspace(0, 2*np.pi, 100)], [2*np.sin(i) for i in np.linspace(0, 2*np.pi, 100)], color='black', linestyle='dotted')
plt.plot([3*np.cos(i) for i in np.linspace(0, 2*np.pi, 100)], [3*np.sin(i) for i in np.linspace(0, 2*np.pi, 100)], color='black', linestyle='dotted')
plt.plot([0, 3], [0, 0], color='black', linestyle='dotted')
plt.plot([0, 3*np.cos(2*np.pi/3)], [0, 3*np.sin(2*np.pi/3)], color='black', linestyle = 'dotted')
plt.plot([0, 3*np.cos(4*np.pi/3)], [0, 3*np.sin(4*np.pi/3)], color='black', linestyle = 'dotted')

plt.scatter([1, 2, 3], [0, 0, 0], color='black')
plt.scatter([np.cos(2*np.pi/3), 2*np.cos(2*np.pi/3), 3*np.cos(2*np.pi/3)], [np.sin(2*np.pi/3), 2*np.sin(2*np.pi/3), 3*np.sin(2*np.pi/3)], color='black')
plt.scatter([np.cos(4*np.pi/3), 2*np.cos(4*np.pi/3), 3*np.cos(4*np.pi/3)], [np.sin(4*np.pi/3), 2*np.sin(4*np.pi/3), 3*np.sin(4*np.pi/3)], color='black')

plt.plot([0, x3], [0, y3], label='$\\vec{p}_{1}$')
plt.plot([0, x4], [0, y4], label='$\\vec{p}_{2}$')
plt.plot([0, x3+x4], [0, y3+y4], label='$\\vec{p}_{1}+\\vec{p}_{2}$')
plt.plot([x4, x3+x4], [y4, y3+y4], color = 'tab:blue', linestyle='dashed')

plt.legend()

plt.gca().set_aspect('equal')
```
